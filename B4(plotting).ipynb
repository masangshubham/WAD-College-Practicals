```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv('/content/heart.csv')
df
```

```python
df.columns
```

```python
#boxplot
sns.boxplot(x='target',y='age',data=df)
```

```python
plt.hist(df['chol'])
plt.xlabel('target')
plt.ylabel('chol')
plt.title('Histogram of target')
plt.show()
```

```python
#line graph
plt.plot(df['age'],df['chol'])
plt.xlabel('age')
plt.ylabel('chol')
plt.title('line graph of age vs chol')
plt.show()
```

```python
sns.lineplot(data=df,x='age',y='chol')
plt.show()
```

```python
sns.regplot(data=df,x='age',y='chol')
```
